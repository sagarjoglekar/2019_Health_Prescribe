```python
drugs_dir =  '../BL_Work/DrugBank_V2/'
drugs = glob.glob(drugs_dir+'*.html')
```

```python
DrugDataDict = {}
for drug in tqdm(drugs):
    key = drug.split('/')[-1].split('.')[0]
    soup = bs(open(drug,'rb').read(),'html.parser')
    DrugDataDict[key] = {'Associations':[] , 'Categories' :[] , 'ATC_Codes':[] , 'Name' : '' }
    try:
        dts = enumerate(soup.find('div', class_='card-content').findAll('dt'))
        dds = soup.find('div', class_='card-content').findAll('dd')
        dtInd = [k for (k,i) in dts if i.text == 'Name' ][0]
        DrugDataDict[key]['Name'] = dds[dtInd].text
    except:
        print "cannot get Name"
    
    try:
        dts = enumerate(soup.find('div', class_='card-content').findAll('dt'))
        dds = soup.find('div', class_='card-content').findAll('dd')
        dtInd = [k for (k,i) in dts if i.text == 'Associated Conditions' ][0]
        associations = [{i.attrs.get('href'): i.text} for i in dds[dtInd].findAll('a')]
        DrugDataDict[key]['Associations'] = associations
    except:
        print "Cannot get Associations"
        
    try:
        dts = enumerate(soup.find('div', class_='card-content').findAll('dt'))
        dds_cat = soup.find('div', class_='card-content').findAll('dd')
        dtInd_cat = [k for (k,i) in dts if i.text == 'Categories' ][0]
        category = [{i.attrs.get('href'): i.text} for i in dds_cat[dtInd_cat].findAll('a')]
        DrugDataDict[key]['Categories'] = category
    except:
        print "cannot get categories"
        
    try:
        dts = enumerate(soup.find('div', class_='card-content').findAll('dt'))
        dds = soup.find('div', class_='card-content').findAll('dd')
        dtInd = [k for (k,i) in dts if i.text == 'ATC Codes' ][0]
        ATC = [{i.attrs.get('href'): i.text} for i in dds[dtInd].findAll('a')]
        DrugDataDict[key]['ATC_Codes'] = ATC
    except:
        print "cannot get ATC Codes"
```

```python
filtered_drug_dict = {}
for k in DrugDataDict:
    if len(DrugDataDict[k]['ATC_Codes']) > 0 : 
        filtered_drug_dict[k] = DrugDataDict[k]
```
