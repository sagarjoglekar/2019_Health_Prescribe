```python
compre_postCode_dict = {}
for name , group in uk_postcode_area.groupby('lad11nm'):
    if name not in compre_postCode_dict:
        compre_postCode_dict[name] = {}
        compre_postCode_dict[name]['areaId'] = list(set(group['lad11cd']))
        compre_postCode_dict[name]['postcodes'] = list(set([k.split(' ')[0] for k in group['pcds']]))
    
    
```
