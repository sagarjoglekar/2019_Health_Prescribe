```python
GP_LSOA_registry = pd.read_csv('../BL_Work/gp-reg-patients-all-lsoa.csv')
```

```python
GP_LSOA_registry.head()
```

```python
GP_LSOA_PATIENT_DIST = {}
for name, group in GP_LSOA_registry.groupby('PRACTICE_CODE'):
    GP_LSOA_PATIENT_DIST[group.iloc[0,0]] = {}
    lsoaIndex = group.iloc[0].index[group.iloc[0].notnull()]
    for i in range(3,len(lsoaIndex)):
        if group.iloc[0,i] == 'NO2011':
            continue
        GP_LSOA_PATIENT_DIST[group.iloc[0,0]][group.iloc[0,i]] = group.iloc[1,i]

```

```python
GP_LSOA_PATIENT_DIST
```

```python
for k in GP_LSOA_PATIENT_DIST:
    denom = float(np.sum(np.array(GP_LSOA_PATIENT_DIST[k].values()).astype(np.float)))
    for j in GP_LSOA_PATIENT_DIST[k]:
        GP_LSOA_PATIENT_DIST[k][j] = float( GP_LSOA_PATIENT_DIST[k][j])/denom
            
```

```python
json.dump(GP_LSOA_PATIENT_DIST,open('GP_LSOA_PATIENTSDIST.json','wb'))
```
