```python
import networkx as nx
import json
```

```python
drug_conditions_mapping = json.load(open('Symptom_drug_probabilities.json','rb'))
```

```python
drug_conditions_mapping
```

```python
Graph = nx.Graph()
for k in drug_conditions_mapping:
    for drug in drug_conditions_mapping[k]:
        if float(drug_conditions_mapping[k][drug]) > 0.0: 
            Graph.add_edge(k,drug, weight = drug_conditions_mapping[k][drug])
```

```python
Graph.edges(data=True)
```

```python
nx.write_edgelist(Graph, "drug_sym_bipartite.dat")
```
